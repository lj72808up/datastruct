```python
class Solution:
    def InversePairs(self, data):
        copy = [e for e in data]
        return self.inverseCore(data,copy,0,len(data)-1)
    
    def inverseCore(self,data,copyData,start,end):
        if start==end:
            copyData[start]=data[start]
            return 0
        mid = (start+end)/2
        leftCount = self.inverseCore(data,copyData,start,mid)
        rightCount = self.inverseCore(data,copyData,mid+1,end)
        i = mid
        j = end
        indexCopy = j
        count = 0
        while i>=start and j>=mid+1:
            if data[i]>data[j]:
                copyData[indexCopy] = data[i]
                indexCopy = indexCopy-1
                i = i-1
                count = count+j-(mid+1)-1
            else:
                copyData[indexCopy] = data[j]
                indexCopy = indexCopy-1
                j = j-1
        while i>=start:
            copyData[indexCopy] = data[i]
            indexCopy = indexCopy-1
        while j>=mid+1:
            copyData[indexCopy] = data[j]
            indexCopy = indexCopy-1
        return leftCount + rightCount+count
```
