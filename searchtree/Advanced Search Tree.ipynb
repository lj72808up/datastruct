### 一. 伸展树
#### 1. 伸展树 
  1. 伸展树(splay tree)也是一种搜索树(ST), 相比于AVL的全树平衡, 其实现更为简便, 不需要记录平衡因子和高度  
  2. 伸展树利用了数据局部性:   
     1. 刚被访问的节点, 不久后会再次访问  
     2. 被访问的下一个节点, 可能处于不久之前刚被访问过的某个节点附近   
     
#### 2. 简易伸展树   
  1. 将查找. 删除, 插入节点后, 把该节点调整称根节点. 深度为n的节点, 可以经过n次调整变成根节点  
  2. 节点的上升过程, 两侧子树不断调整, 此为"伸展". 但最坏情况下, 单次访问的分摊时间复杂度仍可能高达$O(n)$  
  
#### 3. 双层伸展树  
  为克服简易伸展调整策略的缺陷, 每次都将节点上升两层(进行2次调整)

### 二. B树
#### 1. 分级存储  
  1. 现代存储器大多以分级存储, 如内存与外存(磁盘)形成的二级存储.   
  2. 且对于磁盘而言, 连续读取物理地址连续的1000个字节, 和只读取几个字节的时间消耗差不多
  
#### 2. 多路查找树  
  1. 从BST出发, 以k层间隔重组, 可将BST转化为$2^k$路查找树
   <img src="./img/kst.png">
  2. 多路查找树的查找,添加,删除操作虽然效果与BST等同, 但对外存的访问方式已发生本质变化.  此时, 搜索每下降一层, 都以合并后的大节点为单位, 从外存多去一组关键码(不再是单个). 而由于这组关键码物理地址相邻, 其时间消耗与读取单个关键码差异不大
  
#### 3. m阶B树  
1. m阶b树: 指m路平衡查找树.   
  1. 每个节点有不超过m个分支, 不超过m-1个关键码  
  2. 为避免分支数太少, 又规定分支树$n\ge \left\lceil \frac { m }{ 2 }  \right\rceil $, 节点树大于等于$\left\lceil \frac { m }{ 2 }  \right\rceil -1$  
  3. 所有外部节点深度均相同
  
2. b树查找策略 
  1. 首先将大数据集组织成B-树并存放于外存. 查找B树时, 保证其根节点常驻于内存, 且任何时刻只有当前节点被load进内存  
  2. 然后以根节点为当前节点, 逐层深入查找. 若当前节点包含关键码则返回结果, 否则返回逻辑上处于下一层的另一个节点, 并将其load进内存,并作为当前节点.  
可见, 只有在切换和更新当前节点的时候才能发生I/O操作, 而同一节点的关键码查找, 均处于内存之中. 由于节点的关键码范围不大, 因此可采用顺序查找

3. b树关键码插入  
 找到关键码插入位置以后(该位置必处于最底层叶子节点), 若插入后该节点关键码个数仍$\le (m-1)$, 则可直接插入; 否则发生上溢: 将该节点的第$\left\lceil \frac { m }{ 2 }  \right\rceil -1$个关键码上升至其父节点中  
 
4. b树关键码删除  
 找到关键码删除位置后, 若删除后该节点关键码个数仍$n\ge \left\lceil \frac { m }{ 2 }  \right\rceil -1$, 则可直接删除;  
 否则, 发生下溢合并操作: 从父节点选择合适的关键码下降至该节点中, 若父节点不满足关键码个数要求, 则从该节点的兄弟节点中选择关键码提升至父节点, 若提升后导致兄弟节点关键码个数条件不满足, 则可直接将下降的父节点中的关键码和当前节点及其兄弟节点合并为1个大节点.  
 若被删除节点处于根节点, 无父节点进行下溢, 则现将根节点与其直接后继交换后再执行删除

#### 4. m阶b+树
1. m阶b+树每个节点最多m个关键码, m个分支   
2. b+树的内部节点值存储索引, 不存储具体信息. 索引指向孩子节点的最大/最小值所在物理地址  
3. 所有的消息实体, 都存储在b+树的叶子节点  
4. b+树相对于b树的查询优势:   
   1. 因为b+树的内部节点不存储信息, 所以内部节点可包含更多关键码的索引, 因此整个树更矮胖(树高更小)  
   2. b+树查询必须查找到叶子节点，b树只要匹配到即可不用管元素位置，因此b+树查找更稳定（并不慢）  
   3. 对于范围查找来说，b+树只需遍历叶子节点链表即可，b树却需要重复地中序遍历
